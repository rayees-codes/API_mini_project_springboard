**3. Calculate what the highest and lowest opening prices were for the stock in this period.**

```python
# Storing individual stock data into each list, to help in calculations
stock_date = stock_separated[0]  # Date
open_stock = stock_separated[1]  # Open
high_stock = stock_separated[2]  # High
low_stock = stock_separated[3]   # Low
close_stock = stock_separated[4] # Close
trade_volume = stock_separated[7] # Trade volume or No. of shares

# Highest opening price calculation:
sort_high = sorted(open_stock, reverse=True)

print('The highest opening price: ', sort_high[0])

# Lowest opening price calculation:
sort_low = sorted(open_stock, reverse=False)

print('The lowest opening price: ', sort_low[0])
```

**4. What was the largest change in any one day (based on High and Low price)?**


```python
# Finding the difference between high and low price for all the days & finding which day is the largest change 

# Checking the count of items in high stock and low stock to be equal
print(len(high_stock))
print(len(low_stock)) # Both are equal
```

```python
# Checking some sample from high and low stock lists once:
print('High: ', high_stock[:4])
print('Low: ', low_stock[:4])
```

```python
# Calculating the difference between high & low price (high - low = change)

change = []  # Opened a new list to capture the differences for all days

for i in range(len(high_stock)):
    change.append(high_stock[i] - low_stock[i])

print(change[:10]) 

# checking differences for the first ten days
```

```python
# Largest change in any one day

large_change = sorted(change, reverse=True)
print('Largest change in any one day: ', large_change[0])
```

**5. What was the largest change between any two days (based on Closing Price)?**


```python
# Largest change between any two days on closing price: finding the difference between first & second day
change_two = []

for index in range(len(close_stock)-1):
    change_two.append(close_stock[index+1] - close_stock[index])

print(change_two[:10])
# checking differences for the first ten days
```

```python
# Largest change between any two days

large_change_two = sorted(change_two, reverse=True)
print('Largest change between any two days: ', large_change_two[0])
```

**7. (Optional) What was the median trading volume during this year.** 

(Note: you may need to implement your own function for calculating the median.)


```python
# Defining the function to calculate median

def median(your_list):                                 # List as an argument to pass                            
    index_first = 0                                    # First index set as zero                          
    index_last = len(your_list) - 1                    # Last index
    middle_index = (index_first + index_last) / 2      # Finding the middle index
    
    a, b = float(middle_index), int(middle_index)      # assigning variables to each data type (float or integer)
    
    if a == b:                                         # if middle index is integer it passes this function (list count - odd)
        middle_index = int(middle_index)               
        return your_list[middle_index]                 # returns the value at the middle index position
    
    elif a != b:                                       # if middle index is float it passes this function (list count - even)
        left_index = int(middle_index)
        right_index = int(middle_index + 1)
        median = (your_list[left_index] + your_list[right_index]) / 2    # returns the mean value of the two middle values
        return median
        
```
